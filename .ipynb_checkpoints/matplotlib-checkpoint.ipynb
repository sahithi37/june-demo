#piechart
from matplotlib import pyplot as plt
data = [1983984094,847584789,9834903290,234902390,348093823]
l = ['china','India','US','Indonesia','Pakistan']
e = [0.3,0.0,0.0,0.0,0.5]
colors = ['#ccffcc','orange','blue','green','#333300']
plt.pie(data,labels=l,explode=e,shadow=True,startangle=180,wedgeprops = {'edgecolor':"black",'linewidth':2},colors = colors)
plt.title("population Pie chart")
plt.show()
